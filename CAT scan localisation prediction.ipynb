# CAT SCAN LOCATION PREDICTION USING DEEP LEARNING

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

import tensorflow as tf
```

```python
data = pd.read_csv("C:/Users/user/Desktop/capstone/DL/slice_localization_data.csv")
```

Preprocessing

```python
def preprocess_inputs(df):
    df = df.copy()
    
    # Drop patientId column
    df = df.drop('patientId', axis=1)
    
    # Split df into X and y
    y = df['reference']
    X = df.drop('reference', axis=1)
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preprocess_inputs(data)
```

```python
X_train
```

```python
y_train
```

Training

```python
X_train.shape
```

```python
inputs = tf.keras.Input(shape=384,)
x = tf.keras.layers.Dense(128, activation='relu')(inputs)
x = tf.keras.layers.Dense(128, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='linear')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)
model.summary()
model.compile(
    optimizer='adam',
    loss='mse'
)

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=32,
    epochs=100,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
    ]
)
```

```python
plt.plot(history.history["loss"],label="loss")
plt.plot(history.history["val_loss"],label="val_loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.title("Model loss")
plt.show()
```

Results

```python
y_pred = np.squeeze(model.predict(X_test))
```

```python
rmse = np.sqrt(np.mean((y_test - y_pred)**2))
print("RMSE: {:.5f}".format(rmse))

r2 = 1 - (np.sum((y_test - y_pred)**2) / np.sum((y_test - y_test.mean())**2))
print("R^2 Score: {:.5f}".format(r2))
```
